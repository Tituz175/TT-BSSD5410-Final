```python
from PIL import Image, ImageOps
```

```python
img = Image.open("img_50_1_50.jpg")
img = ImageOps.grayscale(img)
img
```

```python
def sepia_grayscale(image, level=0.4):
    # define sepia kernel
    kernel = [
        [0.393, 0.769, 0.189],
        [0.349, 0.686, 0.168],
        [0.272, 0.534, 0.131]
    ]

    # apply convolution
    output = image.copy()
    width, height = output.size
    for x in range(1, width - 1):
        for y in range(1, height - 1):
            conv = 0
            for i in range(3):
                for j in range(3):
                    pixel = image.getpixel((x + i - 1, y + j - 1))
                    conv += pixel * kernel[i][j]

            # adjust sepia level
            conv *= level

            conv = max(0, min(conv, 255))

            output.putpixel((x, y), int(conv))

    return output.convert('RGB')
```

```python
img = sepia_grayscale(img)
img
```

```python
def noise_grayscale(image):
    image = ImageOps.grayscale(image)
    # define noise kernel
    kernel = [
        [0, 1/8, 0],
        [1/8,3/4, 1/8],
        [0, 1/8, 0]
    ]

    # apply convolution
    output = image.copy()
    width, height = output.size
    for x in range(1, width - 1):
        for y in range(1, height - 1):
            conv = 0
            for i in range(3):
                for j in range(3):
                    pixel = image.getpixel((x + i - 1, y + j - 1))
                    conv += pixel * kernel[i][j]

            conv = max(0, min(conv, 255))

            output.putpixel((x, y), int(conv))

    return output.convert('L')
```

```python
img = noise_grayscale(img)
img
```

```python
def blur(image, radius=6):
    # define blur kernel
    kernel = [
        [1/9, 1/9, 1/9],
        [1/9, 1/9, 1/9],
        [1/9, 1/9, 1/9]
    ]

    # apply convolution
    output = image.copy()
    width, height = output.size
    for x in range(radius, width - radius):
        for y in range(radius, height - radius):
            conv = 0
            for i in range(3):
                for j in range(3):
                    pixel = image.getpixel((x + i - radius, y + j - radius))
                    conv += pixel * kernel[i][j]

            conv = int(round(conv))
            output.putpixel((x, y), conv)

    return output
```

```python
img = blur(img)
img
```
